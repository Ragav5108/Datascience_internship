```python
#lineplot
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#bar chart
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values = [10, 20, 15]

plt.bar(categories, values, color='g')
plt.title("Bar Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['X', 'Y', 'Z']
values = [5, 10, 15]

plt.barh(categories, values, color='orange')
plt.title("Horizontal Bar Chart")
plt.show()
```

```python
#scatter
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x, y, color='red')
plt.title("Scatter Plot")
plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt

labels = ['Apples', 'Bananas', 'Cherries']
sizes = [30, 45, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python
#histogram
import matplotlib.pyplot as plt

data = [1,2,2,3,3,3,4,4,5]

plt.hist(data, bins=5, color='blue')
plt.title("Histogram")
plt.show()
```

```python
#subplots
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 3, 4]

plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.title("Plot 1")

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title("Plot 2")

plt.tight_layout()
plt.show()
```

```python
#box plot
import matplotlib.pyplot as plt

data = [7, 15, 13, 12, 9, 11, 12, 10, 15, 20]

plt.boxplot(data)
plt.title("Box Plot")
plt.show()
```

```python
#Area chart
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 3, 2, 5, 4]

plt.fill_between(x, y, color='skyblue')
plt.title("Area Chart")
plt.show()
```

```python
#style plot
import matplotlib.pyplot as plt

plt.style.use('ggplot')  # Use ggplot style

x = [1, 2, 3, 4]
y = [2, 3, 5, 7]

plt.plot(x, y)
plt.title("Styled Plot")
plt.show()
```

```python
#line,color,marker
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y, linestyle='--', color='purple', marker='o')
plt.title("Custom Line Style")
plt.show()
```

```python
#grid line
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.grid(True)
plt.title("Grid Lines")
plt.show()
```

```python
#legend
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [2, 4, 6, 8]
y2 = [1, 3, 5, 7]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.title("Legend Example")
plt.show()
```

```python
#figure size
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))  # Width x Height in inches
x = [1, 2, 3, 4]
y = [1, 2, 3, 4]

plt.plot(x, y)
plt.title("Custom Figure Size")
plt.show()
```

```python
#Axis range
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

plt.plot(x, y)
plt.xlim(0, 5)
plt.ylim(0, 50)
plt.title("Axis Range")
plt.show()
```

```python
#text annotation
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [2, 4, 6]

plt.plot(x, y)
plt.text(2, 4, 'Mid Point', fontsize=12, color='red')
plt.title("Text Annotation")
plt.show()
```

```python
#Tick customization
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

plt.plot(x, y)
plt.xticks([1, 2, 3, 4], ['One', 'Two', 'Three', 'Four'])
plt.yticks([10, 20, 30, 40], ['Low', 'Medium', 'High', 'Top'])
plt.title("Custom Ticks")
plt.show()
```

```python
#Backgroud color
import matplotlib.pyplot as plt

fig = plt.figure()
fig.patch.set_facecolor('lightyellow')

x = [1, 2, 3]
y = [3, 6, 9]

plt.plot(x, y)
plt.title("Background Color")
plt.show()
```

```python
#save plot to file
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [2, 4, 6]

plt.plot(x, y)
plt.title("Save Plot")
plt.savefig("my_plot.png")  # Save as PNG
plt.show()
```

```python
#change font style and size
import matplotlib.pyplot as plt

plt.title("Custom Font", fontsize=14, fontweight='bold', color='blue')
plt.xlabel("X-Axis", fontsize=12)
plt.ylabel("Y-Axis", fontsize=12)

plt.plot([1, 2, 3], [2, 4, 6])
plt.show()
```
