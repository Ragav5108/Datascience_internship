```python
#line plot
import plotly.express as px
df = px.data.gapminder().query("country=='India'")
fig = px.line(df, x='year', y='gdpPercap', title='GDP Per Capita Over Time - India')
fig.show()
```

```python
#bar plot
import plotly.express as px
df = px.data.tips()
fig = px.bar(df, x='day', y='total_bill', color='sex', title='Total Bill by Day and Gender')
fig.show()
```

```python
#scatter plot
import plotly.express as px
df = px.data.iris()
fig = px.scatter(df, x='sepal_width', y='sepal_length', color='species', title='Iris Flower Dimensions')
fig.show()
```

```python
#pie chart
import plotly.express as px
df = px.data.tips()
fig = px.pie(df, names='day', values='total_bill', title='Total Bill Distribution by Day')
fig.show()
```

```python
#Box plot
import plotly.express as px
df = px.data.tips()
fig = px.box(df, x='day', y='total_bill', color='sex', title='Box Plot of Bills by Day and Gender')
fig.show()
```

```python
#Histogram
import plotly.express as px
df = px.data.tips()
fig = px.histogram(df, x='total_bill', nbins=20, title='Histogram of Total Bills')
fig.show()
```

```python
#Bubble chart
import plotly.express as px
df = px.data.gapminder().query("year==2007")
fig = px.scatter(df, x='gdpPercap', y='lifeExp', size='pop', color='continent', hover_name='country', log_x=True, title='GDP vs Life Expectancy (2007)')
fig.show()
```

```python
#Tree map
import plotly.express as px
df = px.data.tips()
fig = px.treemap(df, path=['day', 'time'], values='total_bill', title='Treemap of Bills by Day and Time')
fig.show()
```

```python
#Sun burst chart
import plotly.express as px
df = px.data.tips()
fig = px.sunburst(df, path=['day', 'time'], values='total_bill', title='Sunburst of Bills by Day and Time')
fig.show()
```

```python
#Heat map
import plotly.express as px
df = px.data.tips()
df['tip_percent'] = df['tip'] / df['total_bill']
pivot = df.pivot_table(values='tip_percent', index='day', columns='time')
fig = px.imshow(pivot, text_auto=True, title='Tip Percent Heatmap by Day and Time')
fig.show()
```

```python

#choropleth plot
import plotly.express as px

# Built-in dataset: gapminder
df = px.data.gapminder().query("year == 2007")

fig = px.choropleth(df,
                    locations="iso_alpha",      # ISO 3-letter country codes
                    color="gdpPercap",          # Value to color by
                    hover_name="country",       # Hover label
                    color_continuous_scale="Viridis", # Color scheme
                    title="Global GDP Per Capita (2007)")

fig.show()
```
