```python
#line plot
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = sns.load_dataset("flights")
data = data.pivot(index="month", columns="year", values="passengers")

# Line plot
sns.lineplot(data=data)
plt.title("Line Plot of Flight Passengers")
plt.show()
```

```python
#Bar plot
tips = sns.load_dataset("tips")
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Bar Plot of Total Bill per Day")
plt.show()
```

```python

#count plot
tips = sns.load_dataset("tips")
sns.countplot(x="day", data=tips)
plt.title("Count of Records per Day")
plt.show()
```

```python
#box plot
tips = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box Plot of Total Bill per Day")
plt.show()
```

```python
#violin plot
tips = sns.load_dataset("tips")
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Violin Plot of Total Bill per Day")
plt.show()
```

```python
#strip plot
tips = sns.load_dataset("tips")
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
plt.title("Strip Plot of Total Bill per Day")
plt.show()
```

```python
#swarm plot

tips = sns.load_dataset("tips")
sns.swarmplot(x="day", y="total_bill", data=tips)
plt.title("Swarm Plot of Total Bill per Day")
plt.show()
```

```python

#Histogram
tips = sns.load_dataset("tips")
sns.displot(tips["total_bill"], kde=False)
plt.title("Histogram of Total Bills")
plt.show()
```

```python
#KDE plot / Density plot
tips = sns.load_dataset("tips")
sns.kdeplot(data=tips["total_bill"], fill=True)
plt.title("KDE Plot of Total Bills")
plt.show()
```

```python
#heat map
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")
data = flights.pivot(index="month", columns="year", values="passengers")
sns.heatmap(data, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Heatmap of Passengers")
plt.show()
```

```python
#pair plot
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
sns.pairplot(iris, hue="species")
plt.suptitle("Pair Plot of Iris Dataset")
plt.show()
```

```python
#join plot
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter")
plt.suptitle("Joint Plot of Total Bill and Tip")
plt.show()
```

```python
#Face Grid
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
g = sns.FacetGrid(tips, col="sex")
g.map(sns.histplot, "total_bill")
plt.suptitle("Facet Grid by Sex")
plt.show()
```
